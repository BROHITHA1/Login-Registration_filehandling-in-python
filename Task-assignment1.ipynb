```python
def register():
    db=open("logindata.txt",'r')
    username=input('Username : ')
    password=input('Password : ')
    d=[]
    f=[]
    for  i in db:
        a,b=i.split(',')
        a=a.strip()
        b=b.strip()
        d.append(a)
        f.append(b)
    userlist=dict(zip(d,f))
    if not check_username(username):
        print('Username parameters not satisfied')
        register()
    else:
        if not password_check(password):
            print('Password parameters not satisfied')
            register()
        elif username in d:
            print('Username Exists! , Login')
            register()
        else:
            db=open("logindata.txt",'a')
            db.write(username+" ,"+password+"\n")
            print('Registration Successful')
#register()
def check_username(username):
    ch=''
    at=0
    dot=0
    val=True
    if '@' and '.' in username:
        ch=username[0]
        at=username.find('@')
        dot=username.find('.')
        if((ch >= 'a' and ch <= 'z') or (ch >= 'A' and ch <= 'Z')):
            if at==dot-1:
                print("@ and . shouldn't be adjacent")
                #print(1)
                val=False
                register()
        else:
            print("Username should start with a alphabet")
            #print(2)
            val=False
            register()
    else:
        print("Username must have '@' and '.'")
        val=False
        register()
    return val
def password_check(password):
    SpecialSym =['$', '@', '#', '%','!','^','&','*','(',')','.','_','-',':',';','?','/']
    val = True
      
    if len(password)<6:
        print('length should be at least 6')
        val = False
          
    if len(password)>=15:
        print('length should be not be greater than 15')
        val = False
          
    if not any(char.isdigit() for char in password):
        print('Password should have at least one numeral')
        val = False
          
    if not any(char.isupper() for char in password):
        print('Password should have at least one uppercase letter')
        val = False
          
    if not any(char.islower() for char in password):
        print('Password should have at least one lowercase letter')
        val = False
          
    if not any(char in SpecialSym for char in password):
        print('Password should have at least one of the symbols $@#')
        val = False
    return val

def access():
    db=open("logindata.txt",'r')
    username=input('Username : ')
    password=input('Password : ')
    if not len(username or password)<1:
        d=[]
        f=[]
        for  i in db:
            a,b=i.split(',')
            a=a.strip()
            b=b.strip()
            d.append(a)
            f.append(b)
        userlist=dict(zip(d,f))
        try :
            if userlist[username]:
                #print(userlist[username])
                try:
                    if password==userlist[username]:
                        #print(password)
                        print('Login Successful !')
                    else:
                        print('Username or Password incorrect')
                        access()
                except:
                    print('Incorrect Username or Password ')
                    access()
            else:
                print("Username doesn't exist- Kindly Verify / Register")
                register()
        except:
            print("Username  or Password doesn't exist- Kindly Register !")
            register()
def fg():
    db=open("logindata.txt",'r')
    username=input('Username : ')
    ind=0
    pd=[]
    d=[]
    f=[]
    for  i in db:
        a,b=i.split(',')
        a=a.strip()
        b=b.strip()
        d.append(a)
        f.append(b)
    userlist=dict(zip(d,f))
    #print(d)
    ind=d.index(username)
    #print(ind)
    #print(userlist.values())
    pd=list(userlist.values())
    print(f'Password for {username} is {pd[ind]}')

def home(option=None):
    option=input('Login-->1 | Signup-->2 | Forgot Password-->3 :')
    option.lower()
    if option =='1':
        access()
    elif option=='2':
        register()
    elif option =='3':
        fg()
    else:
        print('Please select a valid option')
home()
    
```
